```python
iris_2.drop_duplicates(keep=('first'), inplace = True)
```

```python
stats = iris_2.describe().transpose()
stats['IQR'] = stats['75%'] - stats['25%']
stats
```

```python
outliers = pd.DataFrame(columns=iris_2.columns)
 
for col in stats.index:
    iqr = stats.at[col,'IQR']
    cutoff = iqr * 1.5
    lower = stats.at[col,'25%'] - cutoff
    upper = stats.at[col,'75%'] + cutoff
    results = iris_2[(iris_2[col] < lower) | 
                   (iris_2[col] > upper)].copy()
    results['Outlier'] = col
    outliers = outliers.append(results)
outliers
```

```python
outliers.value_counts().sum()
```

```python
index_to_drop=outliers.index
index_to_drop
```

```python
iris_2=iris_2.drop(index=index_to_drop)
```
